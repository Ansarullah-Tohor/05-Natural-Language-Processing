```python
#import tensorflow
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# this helps to assigning a unique number to each word
tokenizer=Tokenizer()
tokenizer.fit_on_texts([book])
```

```python
# see the number of unique words
tokenizer.word_index
```

```python
#check the length
vocabulary_size=len(tokenizer.word_index)+1
```

```python
vocabulary_size
```

```python
# tokenize the book as sentences and changes it into numbers
import nltk
nltk.download('punkt')
nltk.download("punkt_tab")

from nltk.tokenize import sent_tokenize

for sentence in sent_tokenize(book):
    print(tokenizer.texts_to_sequences([sentence]))


```

#### input and output pattern in a next word predictor model
Screenshot 2025-10-02 193834.png

```python
#here all the output in double list
# to solve this problem we have to use [0] to get the output in single list
input_sequence=[]
for sentence in sent_tokenize(book):
    tokenized_sentense=tokenizer.texts_to_sequences([sentence])[0]
    #create a ngram of each tokenized sentense
    for i in range(1,len(tokenized_sentense)):
        ngram_sequence=tokenized_sentense[:i+1]
        input_sequence.append(ngram_sequence)
```

#### Idea behind the code is to create ngrams for each sentence and then append it to the input_sequence

**1.Idea**

The model learns:
👉 Given some words (context), what is the most likely next word?

So, input = sequence of tokens (words or subwords),
output = the next token.


**2.Input–Output Pattern
Example sentence:**

"I love natural language processing"

If we use n-gram or sliding window style:

Input: "I" → Output: "love"

Input: "I love" → Output: "natural"

Input: "I love natural" → Output: "language"

Input: "I love natural language" → Output: "processing"

```python
input_sequence
```

```python
# check the max length of input sequences that will help us to padding
max_lenght=max([len(seq) for seq in input_sequence])
```

```python
max_lenght
```

```python
#import pad_sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequence=pad_sequences(input_sequence,maxlen=max_lenght,padding='pre')
padded_input_sequence
```

# undertanding of how input and output can be separate

[  0,   0,   0, ...,   0,  13,  17], here last number 17 is output and all other numbers are input.
for all the row same logic

```python
# take last number of each sequence as output
x=padded_input_sequence[:,:-1]  ## takes all row but all column without last column which is output
y=padded_input_sequence[:,-1]  ## takes all row and just last column which is output
```

### Is it regression or classification task?
Answer: it is a classification task bacause all number that represent words is discrete. It is and multi class classification.


```python
y = np.array(y).flatten()

# bacause above code giving y as shape of (1496,497,497) so we need to convert it into (1496,497)
#from tensorflow.keras.utils import to_categorical
#y=to_categorical(y,num_classes=vocabulary_size)
```

```python
# one hot encoding of y
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=vocabulary_size)
```

```python
y.shape
```

#### in our architecture there will be --->

**embedding layer--> LSTM layer-----> Dense layer**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
embedded_vector_size=100 # 100 feature dimensions
model=Sequential()
model.add(Embedding(vocabulary_size,embedded_vector_size,input_shape=(max_lenght,)))
model.add(LSTM(150))
model.add(Dense(vocabulary_size,activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',patience=20,min_delta=0.000,verbose=1,mode='auto')
```

```python
model.fit(x,y,epochs=20,callbacks=early_stop)
```

#### Now predicting
 step:
 1. take a text
 2. tokenize the text
 3. pad the text
 4. predict


```python
# prediction for one word after the given user given text
text ="He had always"
#tokenize the text
tokenized_text=tokenizer.texts_to_sequences([text])
# pad the text
padded_text=pad_sequences(tokenized_text,maxlen=max_lenght,padding='pre')

pos=np.argmax(model.predict(padded_text))
for word,index in tokenizer.word_index.items():
  if index== pos:
    print(word)
```

## 1️⃣model.predict(padded_token_text)

padded_token_text is your input sequence, already tokenized and padded.

model.predict(...) outputs a probability distribution over the vocabulary for the next word.
Example output (vocab size = 5):

[[0.05, 0.75, 0.10, 0.05, 0.05]]


 Each value is the predicted probability of the corresponding word.**

## 2️⃣ np.argmax(...)

np.argmax returns the index of the maximum value in the array.

This gives the most likely next word according to the model.
From the example above:

np.argmax([[0.05, 0.75, 0.10, 0.05, 0.05]])  # returns 1


Index 1 corresponds to the second word in your tokenizer's vocabulary.

✅ Summary

* model.predict(...) → probability distribution for all words in vocab.

* np.argmax(...) → pick the word with highest probability.

* tokenizer.index_word[pos] → get the actual word.

```python
# prediction
text ="He had always"
# loop for predicting 10 words after the given user text
for i in range (10):
  #tokenize the text
  tokenized_text=tokenizer.texts_to_sequences([text])
# pad the text
  padded_text=pad_sequences(tokenized_text,maxlen=max_lenght,padding='pre')
# store the probability
  pos=np.argmax(model.predict(padded_text))
  # loop to find the pos from tokenizer.word_index
  for word,index in tokenizer.word_index.items():
    # here matching the pos with each index of word_index
    if index== pos:
      text=text +" "+word
      print(text)

```
